```python
running = True
while running:
    mouse_x, mouse_y = pygame.mouse.get_pos()
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if not running:
            break

        # 作る！
        if mouse_x > 205 and mouse_x < 399 and mouse_y > 403 and mouse_y < 489:
            if event.type == pygame.MOUSEBUTTONDOWN:
                if not pannnokazu == 9999999:
                    pannnokazu= pannnokazu + panseisan
                    seisansita = 1

        # 売る！
        if mouse_x > 430 and mouse_x < 528 and mouse_y > 14 and mouse_y < 73:
            if event.type == MOUSEBUTTONDOWN:
                if pannnokazu > 0:
                    kane=  (kane + kanenoryou * pannnokazu)
                    pannnokazu = 0
                    utta = 1
        # 生産力up！
        if mouse_x > 10 and mouse_x < 260 and mouse_y > 100 and mouse_y <160:
            if event.type == MOUSEBUTTONDOWN:
                if panseisan < 20 and kane > (panseisankane - 1):
                        panseisan= panseisan + 1
                        kane = kane - panseisankane
                        panseisankane= panseisankane * 1.5
                        pankatta = 1
                        katta = 1
        # 売値up！
        if mouse_x > 10 and mouse_x < 260 and mouse_y > 170 and mouse_y <230:
            if event.type == MOUSEBUTTONDOWN:
                if kanenoryou < 30 and kane > (urinekane - 1):
                        kanenoryou= kanenoryou + 1
                        kane= kane - urinekane
                        urinekane= urinekane * 1.5
                        urikatta = 1
                        katta = 1
        # トロフィ！
        if mouse_x > 10 and mouse_x < 260 and mouse_y > 240 and mouse_y <300:
            if event.type == MOUSEBUTTONDOWN:
                if torofi == 0 and kane > (torofikane - 1):
                        torofi= 1
                        kane= kane - torofikane
                        torokatta = 1  
                        katta = 1 
                        DisplayKihon()

    if kanenoryou == 30:
        urineurikire = 1

    if panseisan == 20:
        panseisanurikire = 1

    if mouse_x > 10 and mouse_x < 260 and mouse_y > 100 and mouse_y <160:
        panseisanhyouzi2 += 1
        if panseisanhyouzi2 == 1:
            panseisanhyouzi += 1
            panseisanhyouzi3 = 0
            DisplaySeisanKaneMCNum(panseisankane)
        
    else:
        panseisanhyouzi3 += 1
        if panseisanhyouzi3 == 1:
            panseisanhyouzi= 0
            panseisanhyouzi2 = 0
            DisplaySeisanKaneNashiMCNum(panseisankane)
            DisplayKihon()

    if mouse_x > 10 and mouse_x < 260 and mouse_y > 170 and mouse_y <230:
        urinehyouzi2 += 1
        if urinehyouzi2 == 1:
            urinehyouzi += 1
            urinehyouzi3 = 0
            DisplayUrineKaneMCNum(urinekane)
        
    else:
        urinehyouzi3 += 1
        if urinehyouzi3 == 1:
            urinehyouzi= 0
            urinehyouzi2 = 0
            DisplayUrineKaneNashiMCNum(urinekane)
            DisplayKihon()

    if mouse_x > 10 and mouse_x < 260 and mouse_y > 240 and mouse_y <300:
        torofihyouzi2 += 1
        if torofihyouzi2 == 1:
            torofihyouzi += 1
            torofihyouzi3 = 0
            DisplayTorofiKaneMCNum(torofikane)
        
    else:
        torofihyouzi3 += 1
        if torofihyouzi3 == 1:
            torofihyouzi= 0
            torofihyouzi2 = 0
            DisplayKihon()
            if torofi == 0:
                DisplayTorofiKaneNashiMCNum(torofikane)

    if panseisanhyouzi == 1:
        DisplaySeisanKaneNum(panseisankane)

    if urinehyouzi == 1:
        DisplayUrineKaneNum(urinekane)

    if torofihyouzi == 1:
        DisplayTorofiKaneNum(torofikane)

    if seisansita == 1:
        DisplayPanNum(pannnokazu)
        DispalyPanMCNum(pannnokazu)

    if utta == 1:
        DispalyPanMCNum(pannnokazu)
        DisplayKaneNum(kane)
        DisplayPanNum(pannnokazu)
        DisplayKaneMCNum(kane)
    else:
        if katta == 1:
            DispalyPanMCNum(pannnokazu)
            DisplayKaneNum(kane)
            DisplayPanNum(pannnokazu)
            DisplayKaneMCNum(kane)
            if pankatta == 1:
                DisplaySeisanKaneMCNum(panseisankane)
            if urikatta == 1:
                DisplayUrineKaneMCNum(urinekane)
            if torokatta == 1:
                DisplayTorofiKaneMCNum(torofikane)

    seisansita = 0
    utta = 0
    katta = 0
    pankatta = 0
    urikatta = 0
    torokatta = 0
    
    pygame.display.flip()
pygame.quit()
```
